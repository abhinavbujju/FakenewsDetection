```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data = data.drop('id',axis=1)
```

```python
l=data['author'].nunique()
print('Total unique author are : ',l)
```

```python
c = data['author'].value_counts()
for name, count in c.items():
    print('author', name, 'has repeated', count, 'times')
```

```python
c=data['label'].value_counts()
for name, count in c.items():
    print(f"{name}: {count} times")
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
x = data.drop('label',axis=1)
y = data['label']
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
# Fit LabelEncoder on 'author' in the training data
x['author'] = label_encoder.fit_transform(x['author'])
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer


# Initialize the Tokenizer
tokenizer = Tokenizer()

# Fit the Tokenizer on the entire 'title' data (not just a single column)
tokenizer.fit_on_texts(x_train['title'].astype(str))  # Convert to strings if not already

# Convert text to integer sequences for 'title' in both training and testing data
x_train['title'] = tokenizer.texts_to_sequences(x_train['title'].astype(str))
x_test['title'] = tokenizer.texts_to_sequences(x_test['title'].astype(str))

# Fit the Tokenizer on the entire 'text' data (not just a single column)
tokenizer.fit_on_texts(x_train['text'].astype(str))

# Convert text to integer sequences for 'text' in both training and testing data
x_train['text'] = tokenizer.texts_to_sequences(x_train['text'].astype(str))
x_test['text'] = tokenizer.texts_to_sequences(x_test['text'].astype(str))
```

```python
print(x_train)
print(x_test)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Define the maximum sequence length you want (adjust as needed)
max_sequence_length = 1000  # Example: Set to your desired sequence length

# Pad the sequences to a fixed length
x_train['title'] = pad_sequences(x_train['title'], maxlen=max_sequence_length, padding='post', truncating='post')

x_train['text'] = pad_sequences(x_train['text'], maxlen=max_sequence_length, padding='post', truncating='post')

x_test['title'] = pad_sequences(x_test['title'], maxlen=max_sequence_length, padding='post', truncating='post')

x_test['text'] = pad_sequences(x_test['text'], maxlen=max_sequence_length, padding='post', truncating='post')
```

```python
print(x_train)
print(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
x_train.shape
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model
logistic_regression_model = ExtraTreeClassifier()

# Train the model on the training data
logistic_regression_model.fit(x_train, y_train)

# Make predictions on the test data
y_pred = logistic_regression_model.predict(x_test)

# Calculate accuracy
acc = accuracy_score(y_test, y_pred)

print(f'ExtraTreeClassifier:\nAccuracy: {acc:.4f}')
```
